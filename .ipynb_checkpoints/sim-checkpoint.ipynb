# ICA

## Generate Unmixing matrix

```python
from sklearn.decomposition import FastICA, PCA
import pandas as pd
import numpy as np
import csv

reload(sim)

f = data.fs['obse']
fcal = tools.window_sample(data.fs['obse'], data.time_act['calib'])

# Perform ICA
ica = FastICA(n_components=3)
ica.fit(fcal)

data.save_data('unmixing', ica.components_)

param = ica.get_params()

data.save_csv('ica_param', param)
    
param
```

## Unmix force

```python
reload(forseps)
reload(splt)

# Unmix forces
pd.options.display.max_rows = 6000
data.S = forseps.transform(ica, data.fs)
#print data.S

# Write the data to .csv
fileName = "ICA_params_unsorted.csv"
f = open(fileName, 'w')
f.seek(0)
f.write(str(data.S))
f.close()

# Plot unmixed force
#fig, ax = splt.plot3(data.S, 'obse', label = 'auto', unit = 'N', start = 8, end = '')
#splt.plot_spans(ax, data.time_act)
#fig.savefig(report_path + data.desc + "-S.pdf", format='pdf')
#splt.plot3_cmp(ax, data.S, 'aero', label = 'auto')
#splt.plot3_cmp(ax, data.S, 'inte', label = 'auto')
#splt.plot_spans(ax, data.time_act)
#fig.savefig(report_path + data.desc + "-S-comps.pdf", format='pdf')
```

## Determine the most correlated signal component with hover part to be filtered

```python
reload(tools)
# Determine correlated component
S_pre = tools.window_sample(data.S['obse'], [data.time_act['hover'][0]+3, data.time_act['hover'][1]])
S_post = tools.window_sample(data.S['obse'], [data.time_act['calib'][0], data.time_act['calib'][1]])

# Match signal length
d1, d2 = tools.match_length(S_pre, S_post)

# Perform fft
D1 = data.fft(d1)
D2 = data.fft(d2)

# Measure difference in FFT
DE = (abs(D2) - abs(D1))**2
comp_rmse = np.sqrt(DE.mean())
print comp_rmse

# Save comp_rmse
data.save_data('comp_rmse', comp_rmse)

# Find the minimum error
print 'Component with minimum RMS ' + comp_rmse.idxmin()

# Update mixing matrix with filtered component
ica.mixing_filt_ = np.linalg.inv(ica.components_)
ica.mixing_filt_[:, list(comp_rmse).index(comp_rmse.min())] = 0
print ica.mixing_filt_

data.save_data('mixing_filt', ica.mixing_filt_)
```

```python
# Plot time series
fig, ax = splt.plot3(d1, 'hover', label = 'auto', unit = 'N', start = '', end = '')
splt.plot3_cmp(ax, d2, 'calib', label = 'auto')
fig.savefig(report_path + data.desc + "-S-separ.pdf", format='pdf')

# Plot f series
fig, ax = splt.plotF(D1, 'hover', label = 'auto', unit = 'N')
splt.plotF_cmp(ax, abs(D2), 'calib', label = 'auto')
fig.savefig(report_path + data.desc + "-Sf-separ.pdf", format='pdf')
```
